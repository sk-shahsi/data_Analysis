```python
import pandas as pd
path = "D:/Data_Analysis/dataclining.xlsx"
df= pd.read_excel(path)
print(df)
```

```python
import numpy as np
bins = np.linspace(min(df["length"]),max(df["length"]),4)
```

```python
group_name = ["Low","Medium","High"]
```

```python
df["length_bin"] = pd.cut(df["length"],bins,labels=group_name,include_lowest=True)
```
